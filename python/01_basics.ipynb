### üìù –ü—Ä–∞–∫—Ç–∏—á–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è:

3. –°—Ç–≤–æ—Ä–∏—Ç–∏ —Ü–∏–∫–ª, —â–æ –≤–∏–≤–æ–¥–∏—Ç—å –¥–µ–Ω—å —ñ –Ω–æ–º–µ—Ä —Å–µ–∫—Ç–æ—Ä–∞ –∑ –Ω–æ–º–µ—Ä–æ–º —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É.




### –ó–∞–≤–¥–∞–Ω–Ω—è 1. –í–∏–≤–µ–¥–µ–Ω–Ω—è –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –≤–≤–µ–¥–µ–Ω–æ–≥–æ —á–∏—Å–ª–∞.

````python
# –ó–∞–ø–∏—Ç—É—î–º–æ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –≤–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ, —â–æ –ø–æ–∑–Ω–∞—á–∞—î —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞
# input() –ø–æ–≤–µ—Ä—Ç–∞—î —Ç–µ–∫—Å—Ç–æ–≤–∏–π —Ä—è–¥–æ–∫, —Ç–æ–º—É —Ç—Ä–µ–±–∞ –ø–µ—Ä–µ—Ç–≤–æ—Ä–∏—Ç–∏ –π–æ–≥–æ –Ω–∞ —Ü—ñ–ª–µ —á–∏—Å–ª–æ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é int()
vtraty = int(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ (—á–∏—Å–ª–æ): "))

# –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è –≤–≤–µ–¥–µ–Ω–æ–≥–æ —á–∏—Å–ª–∞ —ñ –≤–∏–≤–æ–¥–∏–º–æ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–æ –¥–æ —Ä—ñ–≤–Ω—è –≤—Ç—Ä–∞—Ç

if vtraty == 0:
    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç –Ω–µ–º–∞—î
    print("–í—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –Ω–µ–º–∞—î.")
elif vtraty < 10:
    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ –Ω–µ–≤–µ–ª–∏–∫—ñ (–º–µ–Ω—à–µ 10)
    print("–í—Ç—Ä–∞—Ç–∏ –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –º—ñ–Ω—ñ–º–∞–ª—å–Ω—ñ.")
elif vtraty < 50:
    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ —Å–µ—Ä–µ–¥–Ω—ñ (–≤—ñ–¥ 10 –¥–æ 49)
    print("–í—Ç—Ä–∞—Ç–∏ –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ —Å–µ—Ä–µ–¥–Ω—ñ.")
elif vtraty < 100:
    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ –≤–µ–ª–∏–∫—ñ (–≤—ñ–¥ 50 –¥–æ 99)
    print("–í—Ç—Ä–∞—Ç–∏ –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –≤–µ–ª–∏–∫—ñ.")
else:
    # –Ø–∫—â–æ –≤—Ç—Ä–∞—Ç–∏ –¥—É–∂–µ –≤–µ–ª–∏–∫—ñ (100 —ñ –±—ñ–ª—å—à–µ)
    print("–í—Ç—Ä–∞—Ç–∏ –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –∫–∞—Ç–∞—Å—Ç—Ä–æ—Ñ—ñ—á–Ω—ñ!")

````

### –ó–∞–≤–¥–∞–Ω–Ω—è 2. –ü—Ä–æ—Ä–∞—Ö–æ–≤–∞–Ω–∏–π —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤.

````python
# –ó–∞–ø–∏—Ç—É—î–º–æ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –≤–≤–µ—Å—Ç–∏ —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –¥–ª—è –ø–µ—Ä—à–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É
# –§—É–Ω–∫—Ü—ñ—è input() –ø–æ–≤–µ—Ä—Ç–∞—î —Ç–µ–∫—Å—Ç, —Ç–æ–º—É –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ –π–æ–≥–æ –Ω–∞ —á–∏—Å–ª–æ (float, —â–æ–± –º–æ–∂–Ω–∞ –±—É–ª–æ –≤–≤–æ–¥–∏—Ç–∏ –¥—Ä–æ–±–æ–≤—ñ —á–∏—Å–ª–∞)
ready1 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –ø–µ—Ä—à–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É: "))

# –ó–∞–ø–∏—Ç—É—î–º–æ —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –¥—Ä—É–≥–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É
ready2 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –¥—Ä—É–≥–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É: "))

# –ó–∞–ø–∏—Ç—É—î–º–æ —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä–µ—Ç—å–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É
ready3 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä–µ—Ç—å–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É: "))

# –û–±—á–∏—Å–ª—é—î–º–æ —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ
# –°—É–º—É—î–º–æ —Ç—Ä–∏ –∑–Ω–∞—á–µ–Ω–Ω—è —ñ –¥—ñ–ª–∏–º–æ –Ω–∞ 3
average_ready = (ready1 + ready2 + ready3) / 3

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω
print("–°–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤:", average_ready)
````

### –ó–∞–≤–¥–∞–Ω–Ω—è 3. –°—Ç–≤–æ—Ä–µ–Ω–æ —Ü–∏–∫–ª, —â–æ –≤–∏–≤–æ–¥–∏—Ç—å –¥–µ–Ω—å —ñ –Ω–æ–º–µ—Ä —Å–µ–∫—Ç–æ—Ä–∞ –∑ –Ω–æ–º–µ—Ä–æ–º —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É.

````python
# –ü—Ä–∏–ø—É—Å—Ç–∏–º–æ, —â–æ –º–∏ –º–∞—î–º–æ 7 –¥–Ω—ñ–≤ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —Ç–∏–∂–¥–µ–Ω—å)
# –Ü –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –¥–Ω—è —î —Å–µ–∫—Ç–æ—Ä —ñ –Ω–æ–º–µ—Ä —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É, —è–∫—ñ –º–∏ –≤–∏–≤–æ–¥–∏–º–æ

# –ö—ñ–ª—å–∫—ñ—Å—Ç—å –¥–Ω—ñ–≤ –¥–ª—è —Ü–∏–∫–ª—É
days_count = 7

# –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ —Ü–∏–∫–ª for –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä—É –¥–Ω—ñ–≤ –≤—ñ–¥ 1 –¥–æ 7 –≤–∫–ª—é—á–Ω–æ
for day in range(1, days_count + 1):
    # –ü—Ä–∏–ø—É—Å—Ç–∏–º–æ, —â–æ –Ω–æ–º–µ—Ä —Å–µ–∫—Ç–æ—Ä–∞ –¥–æ—Ä—ñ–≤–Ω—é—î –Ω–æ–º–µ—Ä—É –¥–Ω—è (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, —Å–µ–∫—Ç–æ—Ä 1 –¥–ª—è –¥–Ω—è 1 —ñ —Ç.–¥.)
    sector_number = day
    
    # –ù–æ–º–µ—Ä —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É –º–æ–∂–Ω–∞ –∑–∞–¥–∞—Ç–∏ —è–∫ –¥–æ–≤—ñ–ª—å–Ω–µ —á–∏—Å–ª–æ, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, day * 10 (–∞–±–æ –±—É–¥—å-—è–∫–∞ –ª–æ–≥—ñ–∫–∞)
    calculation_number = day * 10
    
    # –í–∏–≤–æ–¥–∏–º–æ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é —É –∑—Ä—É—á–Ω–æ–º—É —Ñ–æ—Ä–º–∞—Ç—ñ
    print(f"–î–µ–Ω—å {day}: —Å–µ–∫—Ç–æ—Ä {sector_number}, –Ω–æ–º–µ—Ä —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É {calculation_number}")
````
